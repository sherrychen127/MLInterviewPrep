## Knapsack Problems

The problem says:
- You have items
- Each has a cost/weight
- You have capacity/budget
- You want to maximize or minimize something

#### Unbounded Knapsack
You can use items unlimited times
- coins
- rod cutting
- buy multiple shares
- infinite supply  
State `(remaining_budget)`

```python
def knapsack_unbounded(costs, values, capacity):
    def dp(cap):
        if cap == 0:
            return 0
        res = 0
        for i in range(len(costs)):
            if costs[i] <= cap:
                res = max(res, values[i] + dp(cap - cost[i]))
        return res
    return dp(capacity)
```
