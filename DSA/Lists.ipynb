## LeetCode 311. Sparse Matrix Multiplication

```python
class Solution:
    def multiply(self, mat1, mat2):
        m, k = len(mat1), len(mat1[0])
        n = len(mat2[0])
        
        result = [[0] * n for _ in range(m)]
        
        for i in range(m):
            for t in range(k):
                if mat1[i][t] != 0:
                    for j in range(n):
                        result[i][j] += mat1[i][t] * mat2[t][j]
                        
        return result

#O(m * k * n)
```
