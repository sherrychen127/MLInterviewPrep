## BFS - Breadth First Search
- Shortest path in unweighted graphs
- Level order traversal
- Minimum steps problems

```python
from collections import deque

def bfs(start, graph):
    queue = deque([start])
    visited = set([start])
    while queue:
        node = queue.popleft()
        for neighbor in graph[node]:
            if nei not in visited:
                visited.add(nei)
                queue.append(nei)
```

### BFS with distance

```python
from collections import deque

def bfs(start, graph):
    queue = deque([(start, 0)])  # (node, distance)
    visited = set([start])

    while queue:
        node, dist = queue.popleft()

        for nei in graph[node]:
            if nei not in visited:
                visited.add(nei)
                queue.append((nei, dist + 1))
```

### Topological sort

```python
def topo_sort_dfs(n, adj):
    visited = [0]*n   # 0=unvisited, 1=visiting, 2=done
    res = []

    def dfs(u):
        if visited[u] == 1:
            return False   # cycle
        if visited[u] == 2:
            return True

        visited[u] = 1
        for v in adj[u]:
            if not dfs(v):
                return False
        
        visited[u] = 2
        res.append(u)
        return True

    for i in range(n):
        if not dfs(i):
            return []   # cycle exists

    return res[::-1]
```

## Dijkstra Algorithm

Using Min Heap: $O(ELogV)$
- Each edge relaxation -> heap push $O(logV)$

```python
import heapq as hq
from collections import defaultdict

def dijkstra(n, edges, src):
    #build adjacency list
    adj = defaultdict(lambda:[])
    for (u, v, c) in edges:
        adj[u].append((v, c))

    dist = [float('inf')] * n
    parent = [-1] * n
    queue = [(0, src)]
    dist[src] = 0

    while queue:
        cur_cost, cur_node = hq.heappop(queue)
        if cur_cost != dist[cur_node]:
            continue
        for (neighbor, next_cost) in adj[cur_node]:
            if dist[cur_node] + next_cost < dist[neighbor]:
                dist[neighbor] = dist[cur_node] + next_cost
                parent[neighbor] = cur_node
                hp.heappush(queue, (dist[neighbor], neighbor))
    return dist, parent
        
```

## Floyd Warshall Algorithm

```python
def floyd_warshall(n, edges):
    INF = float('inf')
    
    # distance matrix
    dist = [[INF]*n for _ in range(n)]
    
    # self distance = 0
    for i in range(n):
        dist[i][i] = 0
    
    # initialize edges
    for u, v, w in edges:
        dist[u][v] = w   # directed
        # dist[v][u] = w  # if undirected
    
    # main DP
    for k in range(n):
        for i in range(n):
            for j in range(n):
                dist[i][j] = min(
                    dist[i][j],
                    dist[i][k] + dist[k][j]
                )
    
    return dist
```

## Union Find

1. Find(X)
2. Union(x, y)

Optimization
1. Path Compression
2. Union by Rank/Size

Each node initially points to itself. After union(x, y), the two nodes would share the same root

Union-find = Dynamic graph connectivity
- merging groups
- checking if two nodes are already connected
- building components

```python
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))

    def find(self, x):
        if x != self.parent[x]:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        self.parent[self.find(x)] = self.find(y)
```

```python
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x]) # path compression
        return self.parent[x]

    def union(self, x, y):
        rootX = self.find(x)
        rootY = self.find(y)
        if rootX == rootY:
            return False
            
        if self.rank[rootX] > self.rank[rootY]:
            self.parent[rootY] = rootX
        elif self.rank[rootX] < self.rank[rootY]:
            self.parent[rootX] = rootY 
        else:
            self.parent[rootY] = rootX
            self.rank[rootX] += 1
        return True
```

## LeetCode 200. Number if Islands

```python
class Solution:
    def numIslands(self, grid) -> int:
        N, M = len(grid), len(grid[0])
        def DFS(i, j):
            if i < 0 or j < 0 or i >=N or j >= M or grid[i][j] != '1':
                return
            grid[i][j] = '2'
            DFS(i+1, j)
            DFS(i-1, j)
            DFS(i, j+1)
            DFS(i, j-1)
        count = 0
        for i in range(N):
            for j in range(M):
                if grid[i][j] == '1':
                    DFS(i, j)
                    count += 1
        return count
```

## LeetCode 305. Number if Islands II

```python
class Union:
    def __init__(self, n):
        self.parent = [-1 for _ in range(n)]
        self.rank = [0 for _ in range(n)]
        self.count = 0

    def find(self, node):
        if self.parent[node] == -1:
            return -1
        elif node == self.parent[node]:
            return node
        self.parent[node] = self.find(self.parent[node])
        return self.parent[node]

    def union(self, n1, n2):
        root1 = self.find(n1)
        root2 = self.find(n2)

        if root1 == root2:
            return False
        
        if self.rank[root1] < self.rank[root2]:
            self.parent[root1] = root2
        elif self.rank[root1] > self.rank[root2]:
            self.parent[root2] = root1
        else:
            self.parent[root2] = root1
            self.rank[root1] += 1
        self.count -= 1
        print(f"union:{self.count}")
        return True
    
    def addLand(self, node):
        if self.parent[node] == -1: #must handle duplicate
            self.parent[node] = node
            self.count += 1

    def getNumIsland(self):
        return self.count

    def isLand(self, node):
        if self.parent[node] != -1:
            return True
        return False

class Solution:
    def numIslands2(self, m: int, n: int, positions):
        union = Union(m*n)

        def get_node(x, y):
            # x, y = pos[0], pos[1]
            return n*x + y
        
        answers = []
        for pos in positions:
            node = get_node(pos[0], pos[1])
            union.addLand(node)

            #iterate through the neighbors:
            delta = [[0, 1], [0, -1], [1, 0], [-1, 0]]
            for d in delta:
                npos = [pos[0] + d[0], pos[1] + d[1]]
                if npos[0] < 0 or npos[1] < 0 or npos[0] >= m or npos[1] >= n:
                    continue
                neighbor_node = get_node(npos[0], npos[1])
                if union.isLand(neighbor_node):
                    union.union(node, neighbor_node)
            answers.append(union.count)
        return answers
```
