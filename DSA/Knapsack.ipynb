## Knapsack Problems

The problem says:
- You have items
- Each has a cost/weight 
- You have capacity/budget `c`
- You want to maximize or minimize something, or count ways

> Choose a subset of items under a capacity constraint to optimize something

#### 0/1 Knapsack

Each item can be used at most once
- choose some stocks
- buy each house
- take or leave each item

State `(i, remaining_budget)`
```
dp[i][c] = max value using first i items with capacity c (aka remaining budget)
```
2D DP

```
dp[i][c] = max(
    dp[i-1][c],                   #dont take item i
    dp[i][c-cost[i]] + value[i]   # take item i
)
```

#### Unbounded Knapsack
You can use items unlimited times
- coins
- rod cutting
- buy multiple shares
- infinite supply  
State `(remaining_budget)`

````python
def knapsack_unbounded(costs, values, capacity):
    def dp(cap):
        if cap == 0:
            return 0
        res = 0
        for i in range(len(costs)):
            if costs[i] <= cap:
                res = max(res, values[i] + dp(cap - cost[i]))
        return res
    return dp(capacity)
````
