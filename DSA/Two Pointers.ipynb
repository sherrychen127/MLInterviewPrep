### Opposite Direction Two Pointers
(e.g. Two Sum in sorted array, remove duplicates, container with most water)


````python
def two_pointers_opposite(arr, target):
    left = 0
    right = len(arr) - 1
    while left < right:
        curr = arr[left] + arr[right] # problem specific 
        if cur == target:
            return left, right
        elif cur < target:
            left += 1
        else:
            right -= 1
    return -1
````

## LeetCode 11. Container with Most Water

```
width * min(left_height, right_height)
```
At each step:
1. compute current area
2. move the pointer with the smaller height

````python
    def maxArea(self, height) -> int:
        left, right = 0, len(height)-1
        res = 0
        
        while left < right:
            area = (right - left) * min(height[left], height[right])
            res = max(area, res)
            if height[left] < height[right]:
                left += 1
            else:
                right -=1
        return res
````

### LeetCode 3. Longest Substring Without Repeating Characters

````python
def lengthOfLongestSubstring(self, s: str) -> int:
    last = {}
    l = 0
    ans = 0

    for r in range(len(s)):
        if s[r] in last and last[s[r]] >= l:
            l = last[s[r]] + 1

        last[s[r]] = r
        ans = max(ans, r - l + 1)

    return ans
````
