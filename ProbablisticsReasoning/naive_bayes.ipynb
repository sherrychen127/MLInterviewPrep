We want to classify: 
$$x = (x_1, x_2, ..., x_d)$$

into label: 
$$y \in \{1, ..., K\}$$

We want to compute $P(y|x)$
> probability of class $y$ given features $x$

Using Bayes theorem:
$$P(y|x) = \frac{P(x|y)P(y)}{P(x)}$$

Since $P(x)$ is same for all classes (i.e. probability of observing the feature vector $x$ under the overall data distribution) -> we just compare:
$$P(y|x) \propto P(x|y)p(y)$$

**Q: Why called naive?**  
We assume all features are conditionally independent given the class. 

$$P(x|y) = P(x1, x2, ..., x_d | y) = \prod_i P(x_i | y)$$

The entire learning problem becomes: 
> For each class $y$ and feature $i$ -> learn distribution of feature $x_i$

- Bernoulli NB
- Multinomial NB
- Gaussian NB


discrete features -> classic NB 
continuous features -> Gaussian NB
