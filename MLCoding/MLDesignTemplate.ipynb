````python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
````

````python
class SimpleDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.long)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
        
````

````python
class MLP(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_classes):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_dim, num_classes)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
        
````

````python
def train_model(mode, train_loader, val_loader, num_epochs=5, lr=1e-3):
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    model.to(device)

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0

        for X_batch, y_batch in train_loader:
            X_batch = X_batch.to(device)
            y_batch = y_batch.to(device)

            optimizer.zero_grad()
            logits = model(X_batch)
            loss = criterion(logits, y_batch)
            
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        #validation
        model.eval()
        correct = 0 
        total = 0
        with torch.no_grad():
            for X_batch, y_batch in val_loader:
                X_batch = X_batch.to(device)
                y_batch = y_batch.to(device)

                logits = model(X_batch)
                preds = torch.argmax(logits, dim=1)

                correct += (preds == y_batch).sum().item()
                total += y_batch.size(0)

        print(f"Epoch {epoch+1}: train_loss={running_loss:.4f}, val_acc={correct/total:.4f}")
````

## Ridge Regression
Ridge = linear regression + L2 penalty on weights.


````python
import torch
import torch.nn as nn
import torch.optim as optim

class RidgeRegressionModel(nn.Module):
    def __init__(self, in_features: int):
        super().__init__()
        self.linear = nn.Linear(in_features, 1)

    def forward(self, x):
        return self.linear(x)


X = torch.randn(100, 5)
y = torch.randn(100, 1)

model = RidgeRegressionModel(in_features=5)
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=1e-2)

lambda_l2 = 0.1

for epoch in range(200):
    optimizer.zero_grad()
    pred = model(X)
    mse_loss = criterion(pred, y)

    l2_penalty = torch.sum(model.linear.weight ** 2)
    # l1_penalty = torch.sum(torch.abs(model.linear.weight)) #lasso regression

    loss = mse_loss + lambda_l2 * l2_penalty

    loss.backward()
    optimizer.step()

print("Ridge loss:", loss.item())
````

````python
import torch
import torch.nn as nn
import torch.optim as optim

class LogisticRegressionModel(nn.Module):
    def __init__(self, in_features: int):
        super().__init__()
        self.linear = nn.Linear(in_features, 1)

    def forward(self, x):
        return self.linear(x)  # logits


X = torch.randn(100, 5)
y = torch.randint(0, 2, (100, 1)).float()

model = LogisticRegressionModel(in_features=5)
criterion = nn.BCEWithLogitsLoss()  # sigmoid built in
optimizer = optim.SGD(model.parameters(), lr=1e-2)

for epoch in range(200):
    optimizer.zero_grad()
    logits = model(X)
    loss = criterion(logits, y)
    loss.backward()
    optimizer.step()

with torch.no_grad():
    probs = torch.sigmoid(model(X))
    preds = (probs > 0.5).float()

print("Logistic regression loss:", loss.item())
````

## SVM

````python
import torch
import torch.nn as nn
import torch.optim as optim

class LinearSVM(nn.Module):
    def __init__(self, in_features: int):
        super().__init__()
        self.linear = nn.Linear(in_features, 1)

    def forward(self, x):
        return self.linear(x)  # scores


def hinge_loss(scores, labels):
    # scores: (batch, 1)
    # labels: (batch, 1), values in {-1, +1}
    return torch.mean(torch.clamp(1 - labels * scores, min=0))


X = torch.randn(100, 5)
y = torch.randint(0, 2, (100, 1)).float()
y = 2 * y - 1  # convert {0,1} -> {-1,+1}

model = LinearSVM(in_features=5)
optimizer = optim.SGD(model.parameters(), lr=1e-2)

lambda_l2 = 0.01

for epoch in range(200):
    optimizer.zero_grad()
    scores = model(X)
    loss_hinge = hinge_loss(scores, y)
    l2_penalty = torch.sum(model.linear.weight ** 2)
    loss = loss_hinge + lambda_l2 * l2_penalty
    loss.backward()
    optimizer.step()

with torch.no_grad():
    preds = torch.sign(model(X))

print("SVM loss:", loss.item())
````

## RNN

````python
import torch
import torch.nn as nn

class SimpleRNNModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super().__init__()
        self.rnn = nn.RNN(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True
        )
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        # x: (B, T, input_size)
        out, h_n = self.rnn(x)
        # out: (B, T, hidden_size)
        last_hidden = out[:, -1, :]   # last timestep
        return self.fc(last_hidden)


# Example
x = torch.randn(32, 10, 8)   # batch=32, seq_len=10, input_dim=8
y = torch.randint(0, 3, (32,))

model = SimpleRNNModel(input_size=8, hidden_size=16, num_layers=1, output_size=3)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(50):
    optimizer.zero_grad()
    logits = model(x)
    loss = criterion(logits, y)
    loss.backward()
    optimizer.step()

print("RNN loss:", loss.item())
````

## LSTM

````python
import torch
import torch.nn as nn

class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True
        )
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        # x: (B, T, input_size)
        out, (h_n, c_n) = self.lstm(x)
        last_hidden = out[:, -1, :]
        return self.fc(last_hidden)


# Example
x = torch.randn(32, 10, 8)
y = torch.randint(0, 3, (32,))

model = LSTMModel(input_size=8, hidden_size=16, num_layers=1, output_size=3)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(50):
    optimizer.zero_grad()
    logits = model(x)
    loss = criterion(logits, y)
    loss.backward()
    optimizer.step()

print("LSTM loss:", loss.item())
````

## KNN

````python
import torch

class KNN:
    def __init__(self, k=3):
        self.k = k
        self.X_train = None
        self.y_train = None

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        # X: (num_test, d)
        # X_train: (num_train, d)
        dists = torch.cdist(X, self.X_train)  # (num_test, num_train)

        knn_idx = torch.topk(dists, self.k, largest=False).indices  # nearest k
        knn_labels = self.y_train[knn_idx]  # (num_test, k)

        preds = []
        for labels in knn_labels:
            values, counts = torch.unique(labels, return_counts=True)
            pred = values[counts.argmax()]
            preds.append(pred)

        return torch.stack(preds)


# Example
X_train = torch.randn(20, 4)
y_train = torch.randint(0, 2, (20,))
X_test = torch.randn(5, 4)

knn = KNN(k=3)
knn.fit(X_train, y_train)
preds = knn.predict(X_test)
print("KNN predictions:", preds)
````

## K-means

````python
import torch

class KMeans:
    def __init__(self, n_clusters=3, max_iters=100):
        self.n_clusters = n_clusters
        self.max_iters = max_iters
        self.centroids = None

    def fit(self, X):
        # Random init
        indices = torch.randperm(X.size(0))[:self.n_clusters]
        self.centroids = X[indices]

        for _ in range(self.max_iters):
            # Assign clusters
            dists = torch.cdist(X, self.centroids)  # (N, K)
            labels = torch.argmin(dists, dim=1)

            # Recompute centroids
            new_centroids = []
            for k in range(self.n_clusters):
                cluster_points = X[labels == k]
                if len(cluster_points) == 0:
                    new_centroids.append(self.centroids[k])  # keep old if empty
                else:
                    new_centroids.append(cluster_points.mean(dim=0))

            new_centroids = torch.stack(new_centroids)

            if torch.allclose(self.centroids, new_centroids):
                break

            self.centroids = new_centroids

        return labels

    def predict(self, X):
        dists = torch.cdist(X, self.centroids)
        return torch.argmin(dists, dim=1)


# Example
X = torch.randn(100, 2)

kmeans = KMeans(n_clusters=3, max_iters=50)
labels = kmeans.fit(X)
print("Centroids:\n", kmeans.centroids)
print("Cluster labels:\n", labels[:10])
````

## MLP

````python
import torch
import torch.nn as nn
import torch.optim as optim

class MLP(nn.Module):
    def __init__(self, in_dim, hidden_dims, out_dim):
        super().__init__()

        layers = []
        prev = in_dim

        for h in hidden_dims:
            layers.append(nn.Linear(prev, h))
            layers.append(nn.ReLU())
            prev = h

        layers.append(nn.Linear(prev, out_dim))

        self.net = nn.Sequential(*layers)

    def forward(self, x):
        return self.net(x)


# Example classification
X = torch.randn(200, 10)
y = torch.randint(0, 3, (200,))

model = MLP(10, [64, 32], 3)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

for _ in range(200):
    optimizer.zero_grad()
    logits = model(X)
    loss = criterion(logits, y)
    loss.backward()
    optimizer.step()

print(loss.item())
````

## Gradient Boosting

### Regressor

````python
import torch
import torch.nn as nn
import torch.optim as optim


class WeakRegressor(nn.Module):
    def __init__(self, in_dim, hidden_dim=16):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 1)
        )

    def forward(self, x):
        return self.net(x)


class BoostingRegressor:
    def __init__(self, in_dim, n_estimators=20, learning_rate=0.1, hidden_dim=16):
        self.in_dim = in_dim
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.hidden_dim = hidden_dim
        self.models = []

    def fit(self, X, y, epochs_per_estimator=100, lr=1e-2, verbose=False):
        """
        X: (N, D)
        y: (N, 1)
        """
        pred = torch.zeros_like(y)

        for m in range(self.n_estimators):
            residual = y - pred   # negative gradient for MSE

            model = WeakRegressor(self.in_dim, self.hidden_dim)
            optimizer = optim.Adam(model.parameters(), lr=lr)
            criterion = nn.MSELoss()

            for epoch in range(epochs_per_estimator):
                optimizer.zero_grad()
                update = model(X)
                loss = criterion(update, residual)
                loss.backward()
                optimizer.step()

            with torch.no_grad():
                pred += self.learning_rate * model(X)

            self.models.append(model)

            if verbose:
                total_loss = torch.mean((y - pred) ** 2).item()
                print(f"[Regressor] Estimator {m+1}/{self.n_estimators}, ensemble MSE={total_loss:.6f}")

    def predict(self, X):
        pred = torch.zeros(X.size(0), 1, device=X.device)
        for model in self.models:
            pred += self.learning_rate * model(X)
        return pred
````

````python
torch.manual_seed(42)

N, D = 200, 5
X = torch.randn(N, D)
true_w = torch.tensor([[2.0], [-1.5], [0.5], [0.0], [1.0]])
y = X @ true_w + 0.2 * torch.randn(N, 1)

model = BoostingRegressor(in_dim=D, n_estimators=10, learning_rate=0.2, hidden_dim=8)
model.fit(X, y, epochs_per_estimator=100, lr=1e-2, verbose=True)

pred = model.predict(X)
mse = torch.mean((pred - y) ** 2)
print("Final regressor MSE:", mse.item())
````

### Classifier

````python
import torch
import torch.nn as nn
import torch.optim as optim


class WeakClassifier(nn.Module):
    def __init__(self, in_dim, hidden_dim=16):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 1)
        )

    def forward(self, x):
        return self.net(x)  # output update in logit space


class BoostingBinaryClassifier:
    def __init__(self, in_dim, n_estimators=20, learning_rate=0.1, hidden_dim=16):
        self.in_dim = in_dim
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.hidden_dim = hidden_dim
        self.models = []

    def fit(self, X, y, epochs_per_estimator=100, lr=1e-2, verbose=False):
        """
        X: (N, D)
        y: (N, 1) with values 0 or 1
        """
        logits = torch.zeros_like(y)

        for m in range(self.n_estimators):
            with torch.no_grad():
                probs = torch.sigmoid(logits)
                pseudo_residual = y - probs   # negative gradient

            model = WeakClassifier(self.in_dim, self.hidden_dim)
            optimizer = optim.Adam(model.parameters(), lr=lr)
            criterion = nn.MSELoss()

            for epoch in range(epochs_per_estimator):
                optimizer.zero_grad()
                update = model(X)
                loss = criterion(update, pseudo_residual)
                loss.backward()
                optimizer.step()

            with torch.no_grad():
                logits += self.learning_rate * model(X)

            self.models.append(model)

            if verbose:
                probs = torch.sigmoid(logits)
                bce = nn.BCELoss()(probs, y).item()
                acc = ((probs > 0.5).float() == y).float().mean().item()
                print(f"[Classifier] Estimator {m+1}/{self.n_estimators}, BCE={bce:.6f}, ACC={acc:.4f}")

    def predict_logits(self, X):
        logits = torch.zeros(X.size(0), 1, device=X.device)
        for model in self.models:
            logits += self.learning_rate * model(X)
        return logits

    def predict_proba(self, X):
        return torch.sigmoid(self.predict_logits(X))

    def predict(self, X):
        return (self.predict_proba(X) > 0.5).float()
````

````python
torch.manual_seed(42)

N, D = 300, 6
X = torch.randn(N, D)
w = torch.tensor([[1.5], [-2.0], [0.8], [0.0], [1.2], [-0.7]])
logits_true = X @ w
probs_true = torch.sigmoid(logits_true)
y = torch.bernoulli(probs_true)

clf = BoostingBinaryClassifier(in_dim=D, n_estimators=15, learning_rate=0.2, hidden_dim=8)
clf.fit(X, y, epochs_per_estimator=100, lr=1e-2, verbose=True)

pred = clf.predict(X)
acc = (pred == y).float().mean()
print("Final classifier accuracy:", acc.item())
````
