## LeetCode 70. Climbing Stairs

You are climbing a staircase. It takes n steps to reach the top. Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top. Its actually fibonacci:

$$f(n) = f(n-1) + f(n-2)$$

**Memorization**  
- Time: $O(n)$
- Space: $O(n)$

Without memo -> $O(2^n)$

```python
def climbStairs(self, n: int) -> int:
    memory = {}
    def recursion(n):
        if n in memory:
            return memory[n]
        if n <= 1:
            return 1
        memory[n] = recursion(n-1) + recursion(n-2)
        return memory[n]
    return recursion(n)
```

**Tabular**
It's bottom-up approach


```python
def climbStairs(self, n: int) -> int:
    if n <= 1:
        return 1
    dp = [0] * (n + 1)
    dp[0] = 1
    dp[1] = 1
    for i in range(2, n+1):
        dp[i] = dp[i-2] + dp[i-1]
    return dp[n]
```

**$O(1)$ Space** since we only need last two values

```python
def climbStairs(n):
    if n <= 1:
        return 1
    prev2 = 1
    prev1 = 1
    for _ in range(2, n + 1):
        cur = prev1 + prev2
        prev2 = prev1
        prev1 = cur
    
    return prev1
```

## LeetCode 279. Perfect Squares

Given an integer n, return the least number of perfect square numbers that sum to n.

$$f(n) = 1 + min_{sq \leq n}f(n-sq)$$

**memorization**
- Time: $O(n\sqrt{n})$
- Space: $O(n)$

```python
# memorization
def numSquares(self, n: int) -> int:
    squares = [x**2 for x in range(1, n+1) if x**2 <= n]
    memo = {}
    def backtrack(n):
        if n == 0:
            return 0
        if n in memo: return memo[n]
        res = float('inf')
        for sq in squares:
            if n-sq < 0: break
            res=min(res, 1 + backtrack(n - sq))
        memo[n] = res 
        return  memo[n]
    return backtrack(n)
```

**Bottom-up DP**

Idea: dp[i] = min # of squares to sum to i

```python
def numSquares(self, n: int) -> int:
    #bottom-up 
    squares = [x**2 for x in range(1, n+1) if x**2 <= n]
    dp=[0] + [float('inf')]*n 
    # dp[1] = 1
    for i in range(1, n+1):
        for sq in squares:
            if sq > i: break
            dp[i] = min(dp[i-sq] + 1, dp[i])
    return dp[n]
```

## LeetCode 746. Min Cost Climbing Stairs

You are given an integer array cost where `cost[i]` is the cost of ith step on a staircase. Once you pay the cost, you can either climb one or two steps.


**Memorization**

```python
def minCostClimbingStairs(self, cost) -> int:
    cost += [0]
    memo = {}
    def recursion(n):
        if n in memo: return memo[n]
        if n <= 1:
            return cost[n]
        c = float('inf')
        for step in [1, 2]:
            c = min(c, recursion(n-step)  + cost[n])
        memo[n] = c
        return c
    return recursion(len(cost)-1)
```

**Tabular**  
Think: I'm building a list of already-solved mini-problems so i never redo work. 

`dp[i] = minimum cost to stand on top on step i (after paying cost[i])`

To stand on step `i`, you must come from `i-1` or `i-2`

so:
$$\text{dp}[i] = \text{cost}[i] + min(\text{dp}[i-1], \text{dp}[i-2])$$

Base:
- `dp[0] = cost[0]`
- `dp[1] = cost[1]`

```python
def minCostClimbingStairs(self, cost) -> int:
    n = len(cost)
    if len(cost) <= 2:
        return min(cost)
    prev0, prev1 = cost[0], cost[1]
    for i in range(2, len(cost)):
        cur = cost[i] + min(prev0, prev1)
        prev0, prev1 = prev1, cur
    return min(prev0, prev1)
```

## LeetCode 62. Unique Paths

There is a robot on an m x n grid. The robot is initially located at the top-left corner (i.e., grid[0][0]). The robot tries to move to the bottom-right corner (i.e., grid[m - 1][n - 1]). The robot can only move either down or right at any point in time.

Given the two integers m and n, return the number of possible unique paths that the robot can take to reach the bottom-right corner.


**Memorization**

```python
def uniquePaths(self, m: int, n: int) -> int:
    memo = {}
    def recursion(i, j):
        if (i,j) in memo: return memo[(i,j)]
        if i == 0 or j == 0:
            return 1
        memo[(i,j)] = recursion(i-1, j) + recursion(i, j-1)
        return memo[(i,j)]
    return recursion(m-1, n-1)
```

**Tabular**
$$\text{grid}[i][j] = \text{grid}[i-1][j] + \text{grid}[i][j-1]$$

```python
def uniquePaths(self, m: int, n: int) -> int:
    grid = [[0]*n for _ in range(m)]
    for i in range(m):
        for j in range(n):
            if i == 0 or j == 0:
                grid[i][j] = 1
                continue 
            grid[i][j] = grid[i-1][j] + grid[i][j-1]
    return grid[m-1][n-1]
```

## LeetCode 63. Unique Paths II
An obstacle and space are marked as 1 or 0 respectively in grid. A path that the robot takes cannot include any square that is an obstacle.



```python
def uniquePathsWithObstacles(self, obstacleGrid) -> int:
        m, n = len(obstacleGrid), len(obstacleGrid[0])
        grid = [[0]*n for _ in range(m)]
        if obstacleGrid[0][0] == 1 or obstacleGrid[m-1][n-1] == 1: return 0
        for i in range(m):
            for j in range(n):
                if obstacleGrid[i][j] == 1:
                    grid[i][j] = 0
                    continue 
                if i == 0 and j == 0:
                    grid[i][j] = 1
                elif i == 0:
                    grid[i][j] = grid[i][j-1]
                elif j == 0:
                    grid[i][j] = grid[i-1][j]
                else:
                    grid[i][j] = grid[i-1][j] + grid[i][j-1]
        return grid[m-1][n-1]
```

## LeetCode 64. Minimum Path Sum

Given a m x n grid filled with non-negative numbers, find a path from top left to bottom right, which minimizes the sum of all numbers along its path.

Note: You can only move either down or right at any point in time.

```python
def minPathSum(self, grid) -> int:
    m, n = len(grid), len(grid[0])
    dp = [[0]*n for _ in range(m)]
    for i in range(m):
        for j in range(n):
            if i == 0 and j == 0:
                dp[i][j] =  grid[0][0]
            elif i == 0:
                dp[i][j] = grid[i][j] + dp[i][j-1]
            elif j == 0:
                dp[i][j] = grid[i][j] + dp[i-1][j]
            else:
                dp[i][j] = grid[i][j] + min(dp[i-1][j], dp[i][j-1])
    return dp[m-1][n-1]
```
