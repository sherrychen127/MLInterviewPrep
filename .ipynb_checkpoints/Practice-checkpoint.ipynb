## k-means

```python
import numpy as np
class KMeans:
    def __init__(self, k, max_iter):
        self.k = k
        self.max_iter = max_iter
        
    def fit(self, X): #X: NxD
        n = len(X)
        rand_idx = np.random.choice(np.arange(n), self.k, replace=False)
        centroids = X[rand_idx] #KxD
        for i in range(self.max_iter):
            cluster_assignment = []
            for j in range(n): 
                distances = np.linalg.norm(X[j] - centroids, axis=1)
                cluster_assignment.append(np.argmin(distances))

            for k in range(self.k):
                cur_cluster_idx = np.where(np.array(cluster_assignment) == k)
                cur_cluster = X[cur_cluster_idx]
                if cur_cluster.shape[0] > 0:
                    centroids[k] = np.mean(cur_cluster, axis=0)

            #check for convergence
            if i > 0 and np.array_equal(centroids, previous_centroids):
                break
                
            previous_centroids = np.copy(centroids)
        self.centroids = centroids
        self.cluster_assignment = cluster_assignment

    def predict(self, X):
        cluster_assignment = []
        for i in range(len(X)):
            distances = np.linalg.norm(X[i] - self.centroids, axis=1)
            cluster_assignment.append(np.argmin(distances))
        return cluster_assignment
        
            
        
                
        
        # centroids = X[]
```

## Softmax Logistic Regression

```python
import numpy as np

class SoftmaxRegression:
    
    def __init__(self, lr=0.01, max_iter=1000):
        self.lr = lr
        self.max_iter = max_iter

    def softmax(self, z):
        exp = np.exp(z - np.max(z, axis=1, keepdims=True))
        return exp / np.sum(exp, axis=1, keepdims=True)

    def one_hot(self, y, C):
        N = len(y)
        y_onehot = np.zeros((N, C))
        y_onehot[np.arange(N), y] = 1
        return y_onehot

    def fit(self, X, y):

        N, D = X.shape
        C = len(np.unique(y))

        self.W = np.zeros((D, C))
        self.b = np.zeros(C)

        y_onehot = self.one_hot(y, C)

        for _ in range(self.max_iter):

            z = X @ self.W + self.b
            p = self.softmax(z)

            error = p - y_onehot

            dW = (1/N) * X.T @ error
            db = (1/N) * np.sum(error, axis=0)

            self.W -= self.lr * dW
            self.b -= self.lr * db

    def predict(self, X):
        z = X @ self.W + self.b
        p = self.softmax(z)
        return np.argmax(p, axis=1)
```

## Feedforward

```python
class FeedForward:
    def __init__(self, input_size, hidden_size, output_size, lr=0.01, max_iter=20):
        self.lr = lr 
        self.max_iter = max_iter 
        self.W1 = np.random.rand(input_size, hidden_size)
        self.b1 = np.zeros(hidden_size)
        self.W2 = np.random.rand(hidden_size, output_size)
        self.b2 = np.zeros(output_size)

    def reLU(self, z):
        return np.maximum(0, z)

    def softmax(self, z):
        x = x - np.max(x, axis=1, keepdims=True)
        exp_z = np.exp(z)
        return exp_z/sum(exp_z, axis=1, keepdims=True)

    def one_hot(self, y, num_classes):
        N = len(y)
        one_hot = np.zeros((N, num_classes))
        one_hot[np.arange(N), y] = 1
        return one_hot
        
    def fit(self, X, y):
        n, d = X.shape 

        for i in range(self.max_iter):
            #forward 
            z1 = X @ self.W1 + self.b1 
            a1 = self.reLU(z1)
            z2 = a1 @ self.W2 + self.b2 
            pred = self.softmax(z2)

            #gradient
            error = pred.copy()
            error[np.arange(n), y] -= 1
            # error /= N 
            
            dz2 = error #Nxd
            dW2 = a1.T @ dz2 
            db2 = np.sum(dz2, axis=0)

            dz1 = (dz2 @ W2.T) * (z1 > 0)
            dW1 = X.T @ dz1 
            db1 = np.sum(dz1, axis=0)

            #update
            self.W1 -= self.lr * dW1
            self.b1 -= self.lr * db1 
            self.W2 -= self.lr * dW1
            self.b2 -= self.lr * db2 
        
        
```

    $$\theta = \theta - \eta \nabla_\theta L$$
2. **SGD with Momentum**: adds a velocity term to smooth updates. accelerate in consistent direction and reduce oscillation
    $$v_t = \beta v_{t-1} + \nabla L$$
    $$\theta = \theta - \eta v_t$$


```python
class MomentumSGD:
    def __init__(self, lr, beta):
        self.lr = lr
        self.beta = beta
        self.v = {}
    def step(self, params, grads):
        for k in params:
            if k not in self.v:
                self.v[k] = np.zeros_like(params[k])
            self.v[k] = self.beta * self.v[k] + grads[k]
            self.params[k] -= self.lr * self.v[k]
```

## Convolution

```python
def convolution2D(signal, kernel):
    n = len(signal)
    k = len(kernel)

    padding = k//2
    signal = [0] * padding + signal + [0]*padding
    output = []

    for i in range(n):
        window = signal[i:i+k]
        val = sum(a*b for a, b in zip(window, kernel))
        output.append(val)
    return output
```

```python
import numpy as np

def convolution3D(image, kernel):
    """
    image:  (H, W, C_in)
    kernel: (K_h, K_w, C_in, C_out)
    bias:   (C_out,) or None
    returns: (H_out, W_out, C_out)
    """
    H, W, C_in = image.shape
    Kh, Kw, C_in, C_out = kernel.shape

    pad_h = Kh//2
    pad_w = Kw//2

    #np.pad(array, pad_width, mode)
    image = np.pad(
        image,
        ((pad_h,pad_h),(pad_w,pad_w),(0,0)),
        mode="constant"
    )

    output = np.zeros((H,W,C_out))

    for i in range(H):
        for j in range(W):
            patch = image[i:i+Kh, j:j+Kw, :] #(Kh, Kw, C_in)
            results = patch[...,None] * kernel #(Kh, Kw, C_in, 1) * (Kh, Kw, C_in, C_out)
            output[i,j,:] = np.sum(results, axis=(0,1,2)) # results: (Kh, Kw, C_in, C_out)

    return output
```

```python
import numpy as np

def pooling(image, pool_size=2, stride=2, mode="max"):
    """
    image: (H, W, C)
    pool_size: size of pooling window
    stride: stride of pooling
    mode: "max" or "avg"

    returns: pooled output (H_out, W_out, C)
    """

    H, W, C = image.shape

    H_out = (H - pool_size) // stride + 1
    W_out = (W - pool_size) // stride + 1

    output = np.zeros((H_out, W_out, C))

    for i in range(H_out):
        for j in range(W_out):
            patch = image[
                i*stride:i*stride+pool_size,
                j*stride:j*stride+pool_size,
                :
            ]

            if mode == "max":
                output[i, j, :] = np.max(patch, axis=(0,1))
            elif mode == "avg":
                output[i, j, :] = np.mean(patch, axis=(0,1))

    return output
```

## PCA

```python
import numpy as np

def PCA(X, n_components):
    """
    X: (n_samples, n_features)
    n_components: number of principal components
    """

    # 1. center data
    X_centered = X - np.mean(X, axis=0)

    # 2. covariance matrix
    cov = np.cov(X_centered, rowvar=False)

    # 3. eigen decomposition
    eigenvalues, eigenvectors = np.linalg.eig(cov)

    # 4. sort eigenvectors
    idx = np.argsort(eigenvalues)[::-1]
    eigenvectors = eigenvectors[:, idx]

    # 5. select top components
    components = eigenvectors[:, :n_components]

    # 6. project data
    X_reduced = X_centered @ components

    return X_reduced, components
```

## SVM

```python
import numpy as np

class SVM:
    def __init__(self, learning_rate=0.001, lambda_param=0.01, n_iters=1000):
        self.lr = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.w = None
        self.b = None

    def fit(self, X, y):
        n, d = X.shape
        y_ = np.where(y <= 0, -1, 1)
        self.w = np.zeros(d)
        self.b = 0

        # Gradient descent
        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                condition = y_[idx] * (np.dot(x_i, self.w) - self.b) >= 1
                if condition:
                    self.w -= self.lr * (2 * self.lambda_param * self.w)
                else:
                    self.w -= self.lr * (2 * self.lambda_param * self.w - y_[idx] * x_i)
                    self.b -= self.lr * y_[idx]

    def predict(self, X):
        linear_output = np.dot(X, self.w) - self.b
        return np.sign(linear_output)
```
